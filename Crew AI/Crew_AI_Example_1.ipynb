```python
from openai import OpenAI
import os
from google.colab import userdata
from crewai import Agent, Task, Crew
```

```python
os.environ["OPENAI_API_KEY"]=userdata.get('OPENAI_API_KEY')
client=OpenAI()
```

```python
curriculum_agent = Agent(
    role="Curriculum Analyzer",
    goal="Analyze the syllabus and extract a well defined structure and learning objectives aligning to the trends and standards",
    backstory="Expert in curriculum analysis ,academic design and concurrent trend analysis",
    llm="gpt-4o-mini"
)

instructional_agent = Agent(
    role="Instructional Designer",
    goal="Create a structured weekly teaching plan",
    backstory="Specialist in academic pedagogy.",
    llm="gpt-4o-mini"
)

lab_agent = Agent(
    role="Lab Creator",
    goal="Design hands-on lab exercises aligned with the syllabus",
    backstory="Practical ML engineer turned educator.",
    llm="gpt-4o-mini"
)

assessment_agent = Agent(
    role="Assessment Designer",
    goal="Generate Bloom’s taxonomy aligned assessments",
    backstory="Assessment and evaluation specialist.",
    llm="gpt-4o-mini"
)

presentation_agent = Agent(
    role="Presentation Planner",
    goal="Generate structured slide outlines",
    backstory="Expert academic presenter.",
    llm="gpt-4o-mini"
)
```

```python
syllabus_text = """
Course: Introduction to Model Context Protocol
Duration: 4 weeks
Audience: Computer Science faculty
Focus: MCP, RAG, Autonomous agents
"""

task1 = Task(
    description=f"Analyze this syllabus:\n{syllabus_text}",
    expected_output="Structured analysis of learning objectives, topics, and duration breakdown.",
    agent=curriculum_agent
)

task2 = Task(
    description="Create a weekly teaching plan.",
    expected_output="A detailed week-by-week teaching plan.",
    agent=instructional_agent
)

task3 = Task(
    description="Design practical lab exercises.",
    expected_output="List of hands-on lab exercises aligned with syllabus topics.",
    agent=lab_agent
)

task4 = Task(
    description="Create assessment questions using Bloom's taxonomy.",
    expected_output="Assessment questions categorized by Bloom's taxonomy levels.",
    agent=assessment_agent
)

task5 = Task(
    description="Generate slide outline for each session.",
    expected_output="Structured slide outline per session.",
    agent=presentation_agent
)
crew = Crew(
    agents=[curriculum_agent, instructional_agent, lab_agent, assessment_agent, presentation_agent],
    tasks=[task1, task2, task3, task4, task5],
    verbose=False
)

result = crew.kickoff()
print(result)
```
