```python
def ask_chatgpt(query,model="gpt-4o-mini"):
  response=client.responses.create(model=model,input=query)
  print(response.output_text)

```

```python
def ask_chatgpt_chatbot(query, messages, model="gpt-4o-mini",temp=1.0):
  messages.append({"role": "user", "content": query})
  response = client.chat.completions.create(
      model=model,
      messages=messages,
      temperature=temp
  )
  assistant_message = response.choices[0].message.content
  messages.append({"role": "assistant", "content": assistant_message})
  print(assistant_message)
  return messages
```

```python
conversation_history = []
print("Chatbot started. Type 'quit' to exit.")
while True:
    user_input = input("You: ")
    if user_input.lower() == 'quit':
        break
    conversation_history = ask_chatgpt_chatbot(user_input, conversation_history)
```

```python
def ask_sarcastic_chatbot(query, messages, model="gpt-4o-mini",temp=1.0):
  system_prompt = """You are a sarcastic AI assistant with razor-sharp wit.
  Answer every question with dry humor and playful exaggeration, as if you’re mildly amused by how obvious the answer is.
  However:
  - Never provide incorrect information.
  - Never use profanity.
  - Never attack the user personally.
  - Keep sarcasm clever, not mean.
  Your goal is to entertain while still being accurate and helpful."""


  if not messages or messages[0].get("role") != "system":
      messages.insert(0, {"role": "system", "content": system_prompt})

  messages.append({"role": "user", "content": query})

  response = client.chat.completions.create(
      model=model,
      messages=messages,
      temperature=temp
  )
  assistant_message = response.choices[0].message.content
  messages.append({"role": "assistant", "content": assistant_message})
  print(assistant_message)
  return messages
```

```python
sarcastic_conversation_history = []
print("Sarcastic Chatbot started. Type 'quit' to exit.")
while True:
    user_input = input("You: ")
    if user_input.lower() == 'quit':
        break
    sarcastic_conversation_history = ask_sarcastic_chatbot(user_input, sarcastic_conversation_history)
```

```python
def file_interpret_gpt(file_path, query, model="gpt-4o-mini"):
  try:
    with open(file_path, 'r') as f:
      file_content = f.read()

    combined_input = f"File content:\n{file_content}\n\nUser query: {query}"

    response = client.responses.create(model=model, input=combined_input)
    print(response.output_text)
  except FileNotFoundError:
    print(f"Error: The file at '{file_path}' was not found.")
  except Exception as e:
    print(f"An error occurred: {e}")
```

```python
def file_interpret_chatbot(file_path, query, messages, model="gpt-4o-mini", temp=2.0):
  try:
    with open(file_path, 'r') as f:
      file_content = f.read()


    initial_user_message = f"Here is the content of the file:\n\n{file_content}\n\nNow, regarding this file, my query is: {query}"

    messages.append({"role": "user", "content": initial_user_message})

    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temp
    )
    assistant_message = response.choices[0].message.content
    messages.append({"role": "assistant", "content": assistant_message})
    print(assistant_message)
    return messages

  except FileNotFoundError:
    print(f"Error: The file at '{file_path}' was not found.")
    return messages
  except Exception as e:
    print(f"An error occurred: {e}")
    return messages
```

```python
file_chat_history = []
print("File Chatbot started. Type 'quit' to exit.")
file_path_to_chat = "/content/students_details.txt"

while True:
    user_input = input("You: ")
    if user_input.lower() == 'quit':
        break
    if not file_chat_history:
        file_chat_history = file_interpret_chatbot(file_path_to_chat, user_input, file_chat_history)
    else:
        file_chat_history = file_interpret_chatbot(file_path_to_chat, user_input, file_chat_history)
```
